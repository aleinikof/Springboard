**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transposing the dataframe

transposed_properties = properties.transpose()
transposed_properties.head()
```

```python
# Reseting indices

transposed_properties = transposed_properties.reset_index()
transposed_properties.head()
```

```python
# Assigning the values of the first row to your column headings

transposed_properties.columns = transposed_properties.iloc[0]
transposed_properties.head()
```

```python
# Removing the first row

transposed_properties = transposed_properties.drop(0)
transposed_properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Renaming columns

transposed_properties = transposed_properties.rename(columns={'Unnamed: 0': 'Boroughs', pd.NaT: 'Number'})
transposed_properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melting the dataframe

transformed__properties = pd.melt(frame = transposed_properties, id_vars = ['Boroughs', 'Number'], var_name= 'Date', value_name = 'Value')
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Cheking and changing types

transformed__properties['Value'] = pd.to_numeric(transformed__properties['Value'])
transformed__properties.dtypes
```

```python
transformed__properties.head()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the Number, Value, Date, and Boroughs columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the Boroughs column, and if you find null values, get rid of them however you see fit. 

```python
#Check out columns of the dataframe

transformed__properties['Boroughs'].unique()
```

```python
# Checking for equal number of observations in columns

Boroughs = ['City of London', 'Barking & Dagenham', 'Barnet', 'Bexley',
       'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield',
       'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Haringey',
       'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington',
       'Kensington & Chelsea', 'Kingston upon Thames', 'Lambeth',
       'Lewisham', 'Merton', 'Newham', 'Redbridge',
       'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets',
       'Waltham Forest', 'Wandsworth', 'Westminster']
cleaned__properties = transformed__properties[transformed__properties['Boroughs'].isin(Boroughs)]
print(cleaned__properties['Boroughs'].count())
print(cleaned__properties['Number'].count())
print(cleaned__properties['Date'].count())
print(cleaned__properties['Value'].count())
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular Boroughs? Maybe do a line plot of Date against Value?

```python
# Visualizing the data

cleaned__properties.plot(kind ='line', x = 'Date', y = 'Value')
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Date* column.

```python
# Adding Year do the dataframe

cleaned__properties['Year'] = cleaned__properties['Date'].apply(lambda i: i.year)
cleaned__properties.head()
```

```python
final__properties = cleaned__properties.groupby(by = ['Boroughs', 'Year']).mean()
final__properties
```

```python
final__properties = final__properties.reset_index()
final__properties
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Calculating a ratio

def create_price_ratio(x):
    year_98 = float(x['Value'][x['Year'] == 1998])
    year_18 = float(x['Value'][x['Year'] == 2018])
    ratio = [year_18 / year_98]
    return ratio
```

```python
create_price_ratio(final__properties[final__properties['Boroughs'] == 'Westminster'])
```

```python
# Creating a new dictionary

report = {}
for y in final__properties['Boroughs'].unique():
    area = final__properties[final__properties['Boroughs'] == y]
    report[y] = create_price_ratio(area)
print(report) 
```

```python
# Creating a new dateframe

df = pd.DataFrame(report)
transposed_df = df.transpose()
df_final = transposed_df.reset_index()
df_final.head()
```

```python
df_final.rename(columns={'index':'Area', 0:'2018'}, inplace=True)
df_final.head()
```

```python
greatest = df_final.sort_values(by='2018',ascending=False)
ax = greatest[['Area', '2018']].plot(kind='bar')

ax.set_xticklabels(greatest['Area'])
```
